```python
### WORD2VEC
import warnings
warnings.filterwarnings(action='ignore', category=UserWarning, module='gensim')
from gensim import corpora
from gensim.models import word2vec

### Избавление от шумов, грубая обработка
stopstr = set('<p> </p> <br> <br/> , - / , ( ) . : "'.split())   
stoplist = set('и в на под над для за '.split())
htmlnoiselist = set('<a class= ref= title= width= <small> <br>'.split())


### Скрап сайта
from bs4 import BeautifulSoup
import re
import urllib.request
from urllib.parse import quote
from urllib.error import URLError, HTTPError

def get_page_url(target):
  return "http://www.symbolarium.ru/index.php/{0}".format(quote(target))

def Cleared(str):
    result=str
    for stop in stopstr:
        if stop in result:
            result=result.replace(stop,'')
        elif (result==stop):
            result=""
            return result
    return result

def Noisecheck(str):
    for htmlnoise in htmlnoiselist:
        if htmlnoise in str:
            return False
    return True

def UpperCaseF(str):
    temp = str.lower()
    i=1
    result=''
    for c in temp:
        if i==1:
            result+=c.upper()
            i+=1
        else:
            result+=c
    return result
        

### Переменные
documents=[]
requestlinks=[]

### Обработка сайтов
#Как то улучшить еще? Работает долго очень
#ГЕНЕРАТОРЫ?!
for l in lstr:
    for noun in l.split():
        if (morph.parse(noun)[0].tag.POS=='NOUN'):
            try:
                urlpath = get_page_url(UpperCaseF(noun))
                #Оптимизируем. Не делаем дополнительные запросы по тем же темам.
                if urlpath not in requestlinks:
                    SymbolFile = urllib.request.urlopen(urlpath)
                    requestlinks.append(urlpath)
                else:
                    continue
                print(urlpath)         
            except HTTPError as e:
                #print(e)
                continue
            else:
                SymbolHtml = SymbolFile.read()
                SymbolFile.close()
                soup = BeautifulSoup(SymbolHtml,"lxml")
                links=soup.findAll("p")
                for link in links:
                    documents.append(Cleared(str(link)))
            break
    
### Разделить?!    
   
texts = [[Cleared(word) for word in document.lower().split() if (word not in stoplist)]
         for document in documents if Noisecheck(document)]
#rom pprint import pprint  # pretty-printer
#print(texts)

dictionary = corpora.Dictionary(texts)
f = open('deerwester.dict','w')
dictionary.save('deerwester.dict')
f.close()
f = open('texts.txt','w')
#f.write(texts)
f.close()
#dictionary.load('deerwester.dict')
print(dictionary)
#print(texts)
print("ready")
```

```python
# -*- coding: utf-8 -*-
import scipy.sparse
import numpy as np
import gensim
from sklearn.pipeline import Pipeline
from sklearn.ensemble import ExtraTreesClassifier


class MeanEmbeddingVectorizer(object):
    def __init__(self, word2vec):
        self.word2vec = word2vec
        # if a text is empty we should return a vector of zeros
        # with the same dimensionality as all the other vectors
        self.dim = len(word2vec.itervalues().next())

    def fit(self, X, y):
        return self

    def transform(self, X):
        return np.array([
            np.mean([self.word2vec[w] 
                     for w in words if w in self.word2vec] 
                    or [np.zeros(self.dim)], axis=0) 
            for words in X])


class MyCorpus(object):
    def __iter__(self):
        for line in open('BlockTest\Sbor.txt',encoding="utf8"):
            # assume there's one document per line, tokens separated by whitespace
            yield dictionary.doc2bow(line.lower().split())


class TfidfEmbeddingVectorizer(object):
    def __init__(self, word2vec):
        self.word2vec = word2vec
        self.word2weight = None
        self.dim = len(word2vec.itervalues().next())

    def fit(self, X, y):
        tfidf = TfidfVectorizer(analyzer=lambda x: x)
        tfidf.fit(X)
        # if a word was never seen - it must be at least as infrequent
        # as any of the known words - so the default idf is the max of 
        # known idf's
        max_idf = max(tfidf.idf_)
        self.word2weight = defaultdict(
            lambda: max_idf,
            [(w, tfidf.idf_[i]) for w, i in tfidf.vocabulary_.items()])
        return self

    def transform(self, X):
        return np.array([
                np.mean([self.word2vec[w] * self.word2weight[w] 
                         for w in words if w in self.word2vec] 
                        or 
                        [np.zeros(self.dim)], axis=0) 
            for words in X])          

    
    
w2v=dictionary 
corpus_memory_friendly = MyCorpus()  # doesn't load the corpus into memory!

etree_w2v = Pipeline([
    ("word2vec vectorizer", MeanEmbeddingVectorizer(w2v)),
    ("extra trees", ExtraTreesClassifier(n_estimators=200))])

etree_w2v_tfidf = Pipeline([
    ("word2vec vectorizer", TfidfEmbeddingVectorizer(w2v)),
    ("extra trees", ExtraTreesClassifier(n_estimators=200))])

#print(corpus_memory_friendly)
#for vector in corpus_memory_friendly:  # load one vector into memory at a time
     #print(vector)
#scipy_csc_matrix = gensim.matutils.corpus2csc(corpus)
#print(scipy_csc_matrix)
```
