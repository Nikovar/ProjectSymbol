```python
#!/usr/bin/python
# -*- coding: utf-8 -*-

#import numpy as np
#import pandas as pd
#import pylab as pl
#import math
import collections
#import io
import pymorphy2


#Списки 
lst=[]

def neighbor(pos1,pos2):
    if((pos1=="NOUN") or (pos1=="NPRO")) and ((pos2=="VERB")or(pos2=="ADJF")):
        return True
    if( ((pos1=="VERB") or (pos1=="ADJF")) and ((pos2=="NOUN") or (pos2=="NPRO")) ):
        return True
    return False

class pair:
    def __init__(self,str1,str2):
        self.str1=str1;
        self.str2=str2;
    def __str__(self):
        return self.str1+' '+self.str2


morph = pymorphy2.MorphAnalyzer()
#Лист лишних символов
Unused_chars=[':',',','.','-','\n','?','!']
POS=["NOUN","VERB","ADJF","NPRO"]
#Переменные
path = "D:\\BlockTest\\AutumnDay.txt"
temp = ""
#f=open(path,'r')
f = open(path, encoding="utf8")
for line in f:
    l=line.split('.')
    #print(l)
    for st in l:
        words = st.split()
        #print(len(words))
        for i in range(len(words)):
            #Добавление слова в список слов и избавление от лишних символов
            temp = words[i]
            for char in Unused_chars:
                if char in temp:
                    temp=words[i].replace(char,'')
            pos = morph.parse(temp)[0].tag.POS
            if(pos in POS):
                if(i!=len(words)-1):
                    if(neighbor(pos,morph.parse(words[i+1])[0].tag.POS)):
                        lst.append(pair(temp,words[i+1]))
                
lstr=[]
f.close()
for twos in lst:
    lstr.append(str(twos))
    #print(twos)

print(collections.Counter(lstr))
print('Finish')

```
